```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from collections import Counter
```

```python
file_name = "flights.csv"
flights_data = pd.read_csv(file_name)
flights_data.head()
```

```python
file_name = "airports.csv"
airport_data = pd.read_csv(file_name)
airport_data.head()
```

```python
c_origin = Counter(flights_data.ORIGIN_AIRPORT)
c_destination = Counter(flights_data.DESTINATION_AIRPORT)
len(c_origin)
```

```python
lo = c_origin.most_common(500)
ld = c_destination.most_common(100)
```

```python
cities = []
names = []
values = []
latitude = []
longitude = []
for name,value in lo:
    name = str(name)
    if(len(name) == 3):
        current = airport_data[airport_data.IATA_CODE == name]
        names.append(name)
        cities.append(current.CITY.values[0])
        values.append(int(round(value/365)))
        latitude.append(current.LATITUDE.values[0])
        longitude.append(current.LONGITUDE.values[0])
d = {'city': cities,'name': names, 'flight': values, 'latitude': latitude, 'longitude': longitude}
df = pd.DataFrame(data=d)
df.to_csv("data.csv", encoding='utf-8', index=False)
```
